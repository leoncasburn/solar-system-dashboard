```python
import matplotlib.pyplot as plt

planets = ["Mercury","Venus","Earth","Mars","Jupiter","Saturn","Uranus","Neptune"]
distances= [58000000, 108000000, 150000000, 227900000, 778500000, 1434000000, 2900000000, 4500000000]
moons= [0, 0, 1, 2, 95, 146, 28, 16]
day_length= [1408, 5832, 24, 25, 10, 11, 17, 16]

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].bar(planets, distances)
axes[0].set_title("Distance from Sun (km)")
axes[0].set_ylabel("Distance (km)")

axes[1].bar(planets, moons)
axes[1].set_title("Number of Moons")
axes[1].set_ylabel("Moons")

axes[2].bar(planets, day_length)
axes[2].set_yscale('log')
axes[2].set_title("Length of Day (Hours)")
axes[2].set_ylabel("Day Duration (Hrs)")
axes[2].annotate('Venus rotates\nremarkably slowly!', 
                 xy=(1, 5832), 
                 xytext=(3, 4000),
                 arrowprops=dict(arrowstyle='->'))

axes[0].tick_params(axis='x', rotation=45)
axes[1].tick_params(axis='x', rotation=45)
axes[2].tick_params(axis='x', rotation=45)

axes[0].bar(planets, distances, color='steelblue')
axes[1].bar(planets, moons, color='coral')
axes[2].bar(planets, day_length, color='mediumpurple')

fig.suptitle("Solar System Dashboard", fontsize=16)
plt.tight_layout()
plt.show()
```
